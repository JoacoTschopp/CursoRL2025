**OpenAI Gym: Carpole simple example (random actions)**



```python
import gym
```

```python
env = gym.make("CartPole-v0")

total_reward = 0.0
total_steps = 0
obs = env.reset()

while True:
    action = env.action_space.sample() # take a random action
    obs, reward, done, _ = env.step(action) # execute one step
    total_reward += reward
    total_steps += 1
    if done:
        break

print("Episode done in %d steps, total reward %.2f" % (
    total_steps, total_reward))
```
