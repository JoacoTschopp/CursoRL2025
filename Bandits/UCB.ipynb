Let's import needed packages

```python
import gym_bandits
import numpy as np
import matplotlib.pyplot as plt
import random
from tqdm import tqdm
```

Define the epsilon-Greedy actions (Note: if epsilon=0 it corresponds to the Greedy action)

```python
def get_action(Q, e):
    if random.random() < e:
        return random.randint(0, 9)
    else:
        return np.argmax(Q)
```

We define a function that compute the average of obtained reward over repeated experiments.

```python
def average(tot_rewards):
    avg = []
    for i in range(len(tot_rewards[0])):
        temp = 0
        for j in range(len(tot_rewards)):
            temp += tot_rewards[j][i]
        temp /= len(tot_rewards)
        avg.append(temp)
    return avg
```

We define the UCB function

```python
def ucb(Q, N, t, c):
    vals = Q + c * np.sqrt((np.log(t)) / (N + 1e-8))
    return np.argmax(vals)
```

Main code: define an Environment for 10-armed Bandit whose rewards are normal distributed with standard deviation = 1 and a mean drawn from a normal distribution (0,1). Same as Sutton&Barto book

```python
k = 10  # Number of actions (arms)
iter = 1000 # Number of steps
repeat = 2000 # Number of repetitions
u1 = []  # UCB with c = 1
u2 = []  # UCB with c = 2
e01 = [] # Results for epsilon-Greedy with epsilon = 0.1


np.random.seed(42)
env = gym.make("BanditTenArmedGaussian-v0") # define Environment
```

Run the Greedy algorithm (epsilon = 0.1) several times to average the results

```python
env.reset()
tot_rewards = []
e = 0.1
for _ in tqdm(range(repeat)):
    Q = np.zeros(shape=(k))
    N = np.zeros(shape=(k))
    rewards = []
    for i in range(iter):
        action = get_action(Q, e)
        state, reward, done, info = env.step(action)
        rewards.append(reward)
        N[action] += 1
        Q[action] = Q[action] +  1/N[action] * (reward - Q[action])
    env.reset()
    tot_rewards.append(rewards)

e01 = average(tot_rewards)
```

Run the UCB algorithm with c = 1, several times to average the results

```python
env.reset()
tot_rewards = []
c = 1.0
for _ in tqdm(range(repeat)):
    Q = np.zeros(shape=(k))
    N = np.zeros(shape=(k))
    rewards = []
    for i in range(iter):
        action = ucb(Q, N, i+1, c)
        state, reward, done, info = env.step(action)
        rewards.append(reward)
        N[action] += 1
        Q[action] = Q[action] +  1/N[action] * (reward - Q[action])
    env.reset()
    tot_rewards.append(rewards)

u1 = average(tot_rewards)
```

Run the UCB algorithm with c = 2, several times to average the results

```python
env.reset()
tot_rewards = []
c = 2.0
for _ in tqdm(range(repeat)):
    Q = np.zeros(shape=(k))
    N = np.zeros(shape=(k))
    rewards = []
    for i in range(iter):
        action = ucb(Q, N, i+1, c)
        state, reward, done, info = env.step(action)
        rewards.append(reward)
        N[action] += 1
        Q[action] = Q[action] +  1/N[action] * (reward - Q[action])
    env.reset()
    tot_rewards.append(rewards)

u2 = average(tot_rewards)
```

Finally, we plot the results to compare the Greedy algorithm against the epsilon-Greedy algorithm

```python
plt.plot(u1, color='green', label='c = 1 (UCB)')
plt.plot(u2, color='red', label='c = 2 (UCB)')
plt.plot(e01, color='blue', label='e = 0.1')
plt.xlabel("Steps")
plt.ylabel("Average Reward")
plt.title("Average Reward vs. Steps on 10 Armed Bandit")
plt.legend()
plt.show()
```
